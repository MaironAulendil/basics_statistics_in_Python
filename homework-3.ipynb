Построить частотный график для двух параметров: Freedom и positive affect. Какое это распределение?

```python
n, bins, patches = plt.hist(df['Freedom'], 50, facecolor='b')
plt.xlabel('кол-во',fontsize=20,ha='center')
plt.ylabel('частота',fontsize=20,ha='center')
plt.title('freedom:',fontsize=15,ha='center')
plt.grid(True)
plt.show()
n, bins, patches = plt.hist(df['Positive affect'], 50, facecolor='black')
plt.xlabel('кол-во',fontsize=20,ha='center')
plt.ylabel('частота',fontsize=20,ha='center')
plt.title('positive affect',fontsize=15,ha='center')
plt.grid(True)
plt.show()
```

судя по частотным графикам, распределения величин являются равномерными, причем, оба графика абсолютно идентичны. 

Найти медиану, моду, математическое ожидание, стандартное отклонение,  дисперсию, минимум и максимум, найти выбросы (если есть), размах для колонки Freedom

```python
print('медиана:',statistics.median(df['Freedom']))
print('стандартное отклонение:',max(df['Freedom']))
print('дисперсия:',max(df['Freedom']))
print('минимум:',min(df['Freedom']))
print('максимум:',max(df['Freedom']))
print('размах:',max(df['Freedom'])-min(df['Freedom']))
print('математическое ожидание:',(df['Freedom']).mean())
print('мода:',sts.mode(df['Freedom']).mode)
```

Посмотреть корреляцию между Freedom и positive affect (графически и математически). Графически - plt.scatter, математически - посчитать коэффициент Пирсона. О чем говорит знак коэффициента?

```python
x=df['Freedom']
y=df['Positive affect']
plt.scatter(x,y)
plt.title('граф. корреляция',fontsize=20,ha='center')
plt.xlabel('freedom',fontsize=20,ha='center')
plt.ylabel('positive affect',fontsize=20,ha='center')
plt.grid(True)
plt.show()
df.dropna(inplace=True)
print('коэффициент корреляции пирсона:',sts.pearsonr(x,y))
```

знак коэффициента говорит о том, что коррелляция возрастающая, потому что коэффициент больше нуля.

Нормализовать данные колонки Freedom (централизация + масштабирование)

```python
from sklearn import preprocessing
normaliz =  df['Freedom'] -  df['Freedom'].mean()
print('normalize_mean', normaliz.mean())
min_max = preprocessing.MinMaxScaler(feature_range=(-1,1))
scaled_series = min_max.fit_transform(np.array(normaliz).reshape(-1, 1))
print('min', scaled_series.min())
print('max', scaled_series.max())
```
